```python
concentrations = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```

```python
concentrations.head()
```

```python
concentrations.groupby('genus').aggregate({'sucrose': 'mean', 'alanin': 'mean', 'citrate': 'mean', 'glucose': 'mean', 'oleic_acid': 'mean'})
```

```python
concentrations.groupby('genus').mean()
```

```python
concentrations.groupby('genus').aggregate({'alanin': 'mean', 'alanin': 'min', 'alanin': 'max'})
```

```python
concentrations.groupby('genus').aggregate({'alanin': ['mean', 'min', 'max']}).loc['Fucus'].round(2)
```

```python
concentrations.groupby('group').aggregate({'citrate': 'var'})
```

```python
concentrations.groupby('group').count()
```

```python
concentrations.groupby('group').aggregate(np.ptp)
```
